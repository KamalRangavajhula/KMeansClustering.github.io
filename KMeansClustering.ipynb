```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Importing the packages required for the Project
```

```python
cluster_data = pd.read_csv('C:/Users/kamalsaran/Desktop/textbook/ITMG516/Chapter05DataSet.csv', index_col = 0)
# Reading the dataset and removing the index column
```

```python
cluster_data = cluster_data.drop('Time_in_Community', axis = 1)
cluster_data = cluster_data.drop('Gender', axis =1)
# Dropping columns which are not needed for building the clustering model
```

```python
cluster_data = cluster_data.drop('Id', axis =1)
```

```python
cluster_data.describe()
# Shows the basic summary statistics of the dataset
```

```python
from sklearn.cluster import KMeans
# Importing the KMeans subpackage from the Sckit learn cluster package
kmeans = KMeans(n_clusters=2)
# Specifying the number of clusters required for clustering model
```

```python
kmeans.fit(cluster_data.drop('Working', axis=1 ))
# Fitting the KMeans cluster model to the dataset and removing the cluster attribute from the data
```

```python
kmeans.cluster_centers_
# Gives the detail about the Cluster centers values for each cluster
```

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
    # defining a function to check the accuracy, precision, and confusion matrix
```

```python
cluster_data['Cluster'] = cluster_data['Working'].apply(converter)
# Applying the function to the clustering data model
```

```python
cluster_data.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(cluster_data['Cluster'],kmeans.labels_))
print(classification_report(cluster_data['Cluster'],kmeans.labels_))
# The sckit learn metrics gives the accuracy, precision, and confusion matrix which validates the performance of the model
```
